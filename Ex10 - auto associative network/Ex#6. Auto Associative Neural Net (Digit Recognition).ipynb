```python
#training
weight = []
for i in inp:
    weight.append(np.dot(np.transpose(i),i))
    
print(weight)
```

```python
weights = np.sum(weight,0)
print(weights)
```

```python
step_function = lambda x : 1 if x>0 else -1   
```

```python
def apply(arr):
    for i in range(len(arr)):
        for j in range(arr.shape[-1]):
            arr[i][j] = step_function(arr[i][j])
    return arr
```

```python
#Testing
test_out = []
for i in inp:
    test_out.append(apply(np.dot(i,weights)))

print(test_out)
    
```

```python
def pattern(arr):
    for i in arr:
        print()
        for j in range(len(i)):
            print()
            for k in range(i.shape[1]):
                if(i[j][k] == 1):
                    print('#',end=" ")
                else:
                    print('.',end=" ")
```

```python
pattern(test_out)
```
